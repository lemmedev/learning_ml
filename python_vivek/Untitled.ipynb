```python
class Dog:
    # CLASS OBJECT ATTRIBUTE - SAME FOR ALL INSTANCES OF THE CLASS 
    species = 'mammal'
    def __init__(self, breed, name, spots):
        self.breed = breed
        self.name = name
        self.spots = spots

    # METHODS - FUNCTIONS THAT ARE ASSOCIATED WITH THE CLASS
    def bark(self, number):
        print('Woof! my name is {self.name} and number is {number}') # this will not work because it is not an f-string 
        print(f'Woof! my name is {self.name} and the number is {number}') # this will work because it is an f-string 
        print('Woof! my name is {} and that number is {}'.format(self.name, number)) # this will work because it is using the format method


dog = Dog(breed ='Labrador', name='Buddy', spots=True)
print(dog.breed)
print(dog.name)
print(dog.spots)
print(dog.species)
print(dog.bark(5))
type(dog)
 
```

```python
class Circle:

    # CLASS OBJECT ATTRIBUTE - SAME FOR ALL INSTANCES OF THE CLASS
    pi = 3.1415

    def __init__(self, radius=1):
        self.radius = radius
        self.area = Circle.pi * self.radius ** 2 # we can also use self.pi instead of Circle.pi but it is better to use the class name because it is more clear that we are using a class attribute

    # METHOD
    def get_circumference(self):
        return self.radius * Circle.pi * 2 # we can also use self.pi instead of Circle.pi but it is better to use the class name because it is more clear that we are using a class attribute
    

my_circle = Circle()
circumference = my_circle.get_circumference()
print(circumference, my_circle.area)

```

```python
class Animal:
    def __init__(self):
        print('Animal created') # this is the constructor of the class

    def who_am_i(self):
        print('I am an animal') # this is a method of the class

    def eat(self):
        print('I am eating') # this is a method of the class


class Dog(Animal): # this is how we inherit from a parent class
    def __init__(self):
        Animal.__init__(self) # this is how we call the parent class constructor
        print('Dog created')

    def who_am_i(self):
        print('I am a dog') # this is how we override a method from the parent class

    def bark(self):
        print('Woof!') # this is how we add a new method to the child class


my_dog = Dog()
my_dog.eat() # this is how we call a method from the parent class
my_dog.who_am_i() # this is how we call an overridden method from the child class
my_dog.bark() # this is how we call a method from the child class

```

```python
# Polymorphism - the ability to use a common interface for multiple forms (data types) 
class Dog:
    def __init__(self, name):
        self.name = name

    def speak(self):
        return f'{self.name} says Woof!'

class Cat:
    def __init__(self, name):
        self.name = name

    def speak(self):
        return f'{self.name} says Meow!'
    
niko = Dog('Niko')
felix = Cat('Felix')
print(niko.name + ': ' + niko.speak())
print(felix.name + ': ' + felix.speak())

for pet in [niko, felix]: # this is how we can use polymorphism to call the same method on different objects
    print(type(pet))
    print(pet.speak())

# polymorphism allows us to use the same method name for different classes, and the correct method will be called based on the object that is calling it.
def pet_speak(pet):
    print(pet.speak())

pet_speak(niko) 
pet_speak(felix)
```

```python
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        raise NotImplementedError('Subclass must implement abstract method') # this is how we create an abstract method in python   
    
myanimal = Animal('Fred')
#print(myanimal.name) # this will print the name of the animal
#print(myanimal.speak()) # this will raise an error because we are trying to call an abstract method that has not been implemented in the subclass   

class Dog(Animal):

    def __init__(self, name): # this is the constructor of the subclass 
        Animal.__init__(self, name) 

    def speak(self): # It is not required to call the parent class constructor because it is automatically called when we create an instance of the subclass but it is a good practice to call the parent class constructor explicitly to make it clear that we are calling the parent class constructor and it is also required if we want to pass parameters to the parent class constructor
        return f'{self.name} says Woof!'
    
class Cat(Animal):
    def __init__(self, name): # It is not required to call the parent class constructor because it is automatically called when we create an instance of the subclass but it is a good practice to call the parent class constructor explicitly to make it clear that we are calling the parent class constructor and it is also required if we want to pass parameters to the parent class constructor
        Animal.__init__(self, name) 

    def speak(self): # this is the implementation of the abstract method in the subclass
        return f'{self.name} says Meow!'
    
fido = Dog('Fido')
isis = Cat('Isis')
print(fido.speak())
print(isis.speak())

```

```python
# Special Methods (Magic/Dunder)
class Book:
    def __init__(self, title, author, pages): 
        self.title = title
        self.author = author
        self.pages = pages

    def __str__(self):
        return f"{self.title} by {self.author}"
    
    def __len__(self):
        return self.pages
    
    def __del__(self):
        print('The Book object has been deleted')
        

b = Book('Python Rocks', 'Jose', 200 )
print(b) # Print function essentially, returns string representation of the object
str(b)
len(b)
del(b)

```

```python
class Line:

    def __init__(self, coord1, coord2):
        self.coordinate1 = coord1
        self.coordinate2 = coord2

    def distance(self):
        x1, y1 = self.coordinate1 # Tuple extraction 
        x2, y2 = self.coordinate2
        return ((x2-x1)**2 + (y2-y1)**2) **0.5
    
    def slope(self):
        x1, y1 = self.coordinate1
        x2, y2 = self.coordinate2
        return (y2-y1)/(x2-x1)
    
c1 = (3,2)
c2 = (8,10)
my_line = Line(c1,c2)
my_line.distance()
my_line.slope()
```

```python
class Account():
    
    def __init__(self, owner, balance=0):
        self.owner = owner;
        self.balance = balance

    def deposit(self, dept_amt):
        self.balance = dept_amt + self.balance
        print(f"Added {dept_amt} to the balance")

    def withdrawal(self, wd_amt):
        if self.balance >= wd_amt:
            self.balance = self.balance - wd_amt
            print("Withdrawal accepted")
        else:
            print("Not enough balance")
        
    def __str__(self):
        return f"Owner: {self.owner} \n Balance: {self.balance}"
    

a = Account("Sam", 500)
a.owner
a.balance
a.deposit(100)
print(a)
a.withdrawal(600)
        
    
```
