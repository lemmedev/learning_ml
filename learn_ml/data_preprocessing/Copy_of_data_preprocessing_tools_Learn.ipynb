# Data Preprocessing Tools

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```
dataset = pd.read_csv('Data.csv') # this code uses the read_csv() function from the pandas library to load data from a file named 'Data.csv' into a pandas DataFrame. The resulting DataFrame is then assigned to the variable
X = dataset.iloc[:, :-1] # Matrix of features or Independent variable
Y = dataset.iloc[:, -1] # Dependent variable vector

```

## Taking care of missing data

```
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean') # taking the mean and replacing the nan value.
imputer.fit(X.iloc[:, 1:3]) # In which columns do we want to replace the values. Here it is first and second. It connects imputer to matrix of features
X.iloc[:, 1:3] = imputer.transform(X.iloc[:, 1:3]) # it returns a new updated version of matrix of features of X with replacement from Missing salary and missing age columns.
# Hence we update the matrix of features of X.
```

```
print(X)
```

## Encoding categorical data

### Encoding the Independent Variable

```
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0] )], remainder = 'passthrough')
'''The highlighted code initializes a ColumnTransformer object named ct.
This object is part of scikit-learn's compose module and is used to apply different transformations
to different columns of your dataset. Let's break down its
components: - transformers=[('encoder', OneHotEncoder(), [0])]: This is a list of tuples,
where each tuple defines a transformation. Here, we have one transformation: - 'encoder': This is a
custom name for this specific transformation step. - OneHotEncoder(): This is the transformer itself.
OneHotEncoder converts categorical features (like 'Country' in your X dataframe) into a one-hot
numerical array. For example, if you have 'France', 'Spain', 'Germany', it will create three new
columns, one for each country, with 1 indicating presence and 0 indicating absence. - [0]: This
specifies that the OneHotEncoder should be applied to the column at index 0 of your input data X,
which is currently the 'Country' column. - remainder='passthrough': This argument specifies what
should happen to the columns that are not explicitly mentioned in the transformers list. 'passthrough'
means that any other columns (like 'Age' and 'Salary' in your X dataframe) will be kept in the output
without any modification.'''

X = np.array(ct.fit_transform(X))
```

```
print(X)
```

### Encoding the Dependent Variable

```
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
Y = le.fit_transform(Y)
```

```
print(Y)
```

## Splitting the dataset into the Training set and Test set

```
# feature scaling is applied after the splitting of the dataset because applying feature scaling on the original
# array before the split would cost some info leakage on the test set. We would grab some info from the testset
# which we are not supposed to get because the testset is supposed to be new data with new observation.
# In crux: The essential reason why we should not apply feature scaling before the split it to prevent info
# info leakage on the testset which we are not supposed to have until the training is done.
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 1)
```

## Feature Scaling

```
'''
Feature scaling is important for several reasons, especially in machine learning algorithms:

Preventing Domination by Larger Values: Many machine learning algorithms calculate the distance between data points (e.g., K-Nearest Neighbors, Support Vector Machines, K-Means Clustering). If one feature has a much larger range of values than others (e.g., salary vs. age), its magnitude might disproportionately influence the distance calculations, effectively making the other features seem less important. Feature scaling ensures that all features contribute proportionally to the distance metric.

Faster Convergence for Gradient-Based Optimizers: Algorithms that use gradient descent (like neural networks, logistic regression, and linear regression) converge much faster when the features are scaled. Without scaling, the cost function can have elongated contours, making the optimization path take many small steps, whereas scaled features lead to more spherical contours and a direct path to the minimum.

Regularization: In some regularization techniques (like L1 and L2 regularization), the penalty term is based on the magnitude of the coefficients. If features are not scaled, features with larger values might be penalized more heavily, leading to unfair regularization.

Assumptions of Algorithms: Some algorithms explicitly or implicitly assume that all features are on a similar scale. For example, Principal Component Analysis (PCA) is sensitive to the variance of features, and unscaled features can lead to misleading principal components.

In essence, feature scaling helps algorithms perform better, converge faster, and yield more reliable results by putting all features on a comparable scale.
'''


from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
# feature scaling should not cover dummy variable (like country in the Data.csv).
'''
feature scaling is important for several reasons, especially in machine learning algorithms:

Preventing Domination by Larger Values: Many machine learning algorithms calculate the distance between data points (e.g., K-Nearest Neighbors, Support Vector Machines, K-Means Clustering). If one feature has a much larger range of values than others (e.g., salary vs. age), its magnitude might disproportionately influence the distance calculations, effectively making the other features seem less important. Feature scaling ensures that all features contribute proportionally to the distance metric.

Faster Convergence for Gradient-Based Optimizers: Algorithms that use gradient descent (like neural networks, logistic regression, and linear regression) converge much faster when the features are scaled. Without scaling, the cost function can have elongated contours, making the optimization path take many small steps, whereas scaled features lead to more spherical contours and a direct path to the minimum.

Regularization: In some regularization techniques (like L1 and L2 regularization), the penalty term is based on the magnitude of the coefficients. If features are not scaled, features with larger values might be penalized more heavily, leading to unfair regularization.

Assumptions of Algorithms: Some algorithms explicitly or implicitly assume that all features are on a similar scale. For example, Principal Component Analysis (PCA) is sensitive to the variance of features, and unscaled features can lead to misleading principal components.

In essence, feature scaling helps algorithms perform better, converge faster, and yield more reliable results by putting all features on a comparable scale.


why feature scaling should not cover dummy variable
You're right to ask that! Feature scaling should generally not be applied to dummy variables for a few key reasons:

Loss of Interpretability: Dummy variables (created by one-hot encoding) are inherently categorical and binary, representing the presence (1) or absence (0) of a particular category. Applying scaling (like standardization or normalization) would transform these 0s and 1s into other numerical values, destroying their clear, distinct meaning and making them harder to interpret as categorical indicators.

No Benefit to Algorithms: The purpose of scaling is often to equalize the influence of features with different magnitudes or to aid gradient-based optimization. Dummy variables, by their nature, already have a very limited and consistent range (0 or 1). Scaling them won't typically provide any additional benefit to algorithms; in fact, it can sometimes introduce complexity without improving performance.

Preserving Sparsity: If you have many categorical features, one-hot encoding can lead to a sparse matrix (mostly zeros). Scaling can convert these zeros into non-zero values, destroying the sparsity and potentially increasing computational burden without a corresponding gain.

In summary, dummy variables are already on a 'scale' that is optimal for representing their categorical nature. Applying further scaling would not enhance their utility and could even diminish their clarity and effectiveness in machine learning models.
'''

X_train[:, 3:] = sc.fit_transform(X_train[:, 3:])
X_test[:, 3:] = sc.transform(X_test[:, 3:])


```
